```python
import photutils
from astropy.io import fits
import matplotlib.pyplot as plt 
import matplotlib.colors as mc
import numpy as np 
%matplotlib inline
#%matplotlib widget
```

# photutils: Draw circular Aperture

```python

# first read your fits file 
file = fits.open('data/r.fits') # file name= r.fits
data = file[0].data
#file.info()

# plot it

plt.imshow(data, norm=mc.LogNorm(), origin='lower', cmap='jet')

plt.colorbar()

plt.show()
```

# Circular Apertures at multiple(x,y) Position

```python
import photutils # usef for photometry
from photutils.aperture import CircularAperture # used for circular operture ploting
# define your aperture, give x,y position and radius

aper1 = CircularAperture([618, 580], 300)
aper1.plot(color='red',lw=5)

aper2= CircularAperture([1160, 780], 400)
aper2.plot(color='k',lw=5)
plt.show()
```
